# Header:
## struct io_header;

```python
# HEADER FIELD BYTE FORMATTING:

# >>>(Npart[6])
# calcsize('6I')
# # 24 bytes for six unsigned ints (Npart[6])

# >>>(Massarr[6])
# calcsize('6d')
# # 48 bytes for six doubles (Massarr[6])

# >>>(Time)
# calcsize('1d')
# # 8 bytes for a single double (Time)

# >>>(Redshift)
# calcsize('1d')
# # 8 bytes for a single double (Redshift)

# >>>(FlagSfr)
# calcsize('1i')
# # 4 bytes for an int (FlagSfr)

# >>>(FlagFeedback)
# calcsize('1i')
# # 4 bytes for an int (FlagFeedback)

# >>>(Nall[6])
# calcsize('6i')
# # 24 bytes for six ints (Nall[6])

# >>>(FlagCooling)
# calcsize('1i')
# # 4 bytes for an int (FlagCooling)

# >>>(NumFiles)
# calcsize('1i')
# # 4 bytes for an int (NumFiles)

# >>>(BoxSize)
# calcsize('1d')
# # 8 bytes for a double (BoxSize)

# >>>(Omega0)
# calcsize('1d')
# # 8 bytes for a double (Omega0)

# >>>(OmegaLambda)
# calcsize('1d')
# # 8 bytes for a double (OmegaLambda)

# >>>(HubbleParam)
# calcsize('1d')
# # 8 bytes for a double (HubbleParam)

# >>>(FlagAge)
# calcsize('1i')
# # 4 bytes for an int (FlagAge)

# >>>(FlagMetals)
# calcsize('1i')
# # 4 bytes for an int (FlagMetals)

# >>>(NallHW[6])
# calcsize('6i')
# # 24 bytes for six ints (NallHW[6])

# >>>(flag_entr_ics)
# calcsize('1i')
# # 4 bytes for an int (flag_entr_ics)

# >>>(unused)
# calcsize('1i')
# # 4 bytes for an int (unused)


```

```python

with open("lcdm_gas_littleendian.dat", "rb") as binary_data:
    chunk_size = calcsize('1I')
    block_length = binary_data.read(chunk_size)
    print(block_length)
    block_length_convert = unpack('1I', block_length)
    print('Entire Block Length:', block_length_convert, '\n')
    
    chunk_size = calcsize('6I')
    # 24 bytes for six unsigned ints (Npart[6])
    Npart = binary_data.read(chunk_size)
    print(Npart)
    Npart_convert = unpack('6I', Npart)
    print('Npart[6]:', Npart_convert, '\n')
    
    chunk_size = calcsize('6d')
    # 48 bytes for six doubles (Massarr[6])
    Massarr = binary_data.read(chunk_size)
    print(Massarr)
    Massarr_convert = unpack('6d', Massarr)
    print('Massarr[6]:', Massarr_convert, '\n')
    
    chunk_size = calcsize('1d')
    # 8 bytes for a single double (Time)
    Time = binary_data.read(chunk_size)
    print(Time)
    Time_convert = unpack('1d', Time)
    print('Time:', Time_convert, '\n')
    
    chunk_size = calcsize('1d')
    # 8 bytes for a single double (Redshift)
    Redshift = binary_data.read(chunk_size)
    print(Redshift)
    Redshift_convert = unpack('1d', Redshift)
    print('Redshift:', Redshift_convert, '\n')
    
    chunk_size = calcsize('1i')
    # 4 bytes for an int (FlagSfr)
    FlagSfr = binary_data.read(chunk_size)
    print(FlagSfr)
    FlagSfr_convert = unpack('1i', FlagSfr)
    print('FlagSfr:', FlagSfr_convert, '\n')
    
    chunk_size = calcsize('1i')
    # 4 bytes for an int (FlagFeedback)
    FlagFeedback = binary_data.read(chunk_size)
    print(FlagFeedback)
    FlagFeedback_convert = unpack('1i', FlagFeedback)
    print('FlagFeedback:', FlagFeedback_convert, '\n')
    
    chunk_size = calcsize('6i')
    # 24 bytes for six ints (Nall[6])
    Nall = binary_data.read(chunk_size)
    print(Nall)
    Nall_convert = unpack('6i', Nall)
    print('Nall[6]:', Nall_convert, '\n')
    
    chunk_size = calcsize('1i')
    # 4 bytes for an int (FlagCooling)
    FlagCooling = binary_data.read(chunk_size)
    print(FlagCooling)
    FlagCooling_convert = unpack('1i', FlagCooling)
    print('FlagCooling:', FlagCooling_convert, '\n')
    
    chunk_size = calcsize('1i')
    # 4 bytes for an int (NumFiles)
    NumFiles = binary_data.read(chunk_size)
    print(NumFiles)
    NumFiles_convert = unpack('1i', NumFiles)
    print('NumFiles:', NumFiles_convert, '\n')
    
    chunk_size = calcsize('1d')
    # 8 bytes for a double (BoxSize)
    BoxSize = binary_data.read(chunk_size)
    print(BoxSize)
    BoxSize_convert = unpack('1d', BoxSize)
    print('BoxSize:', BoxSize_convert, '\n')

    chunk_size = calcsize('1d')
    # 8 bytes for a double (Omega0)
    Omega0 = binary_data.read(chunk_size)
    print(Omega0)
    Omega0_convert = unpack('1d', Omega0)
    print('Omega0:', Omega0_convert, '\n')

    chunk_size = calcsize('1d')
    # 8 bytes for a double (OmegaLambda)
    OmegaLambda = binary_data.read(chunk_size)
    print(OmegaLambda)
    OmegaLambda_convert = unpack('1d', OmegaLambda)
    print('OmegaLambda:', OmegaLambda_convert, '\n')
    
    chunk_size = calcsize('1d')
    # 8 bytes for a double (HubbleParam)
    HubbleParam = binary_data.read(chunk_size)
    print(HubbleParam)
    HubbleParam_convert = unpack('1d', HubbleParam)
    print('HubbleParam:', HubbleParam_convert, '\n')

    chunk_size = calcsize('1i')
    # 4 bytes for an int (FlagAge)
    FlagAge = binary_data.read(chunk_size)
    print(FlagAge)
    FlagAge_convert = unpack('1i', FlagAge)
    print('FlagAge:', FlagAge_convert, '\n')
    
    chunk_size = calcsize('1i')
    # 4 bytes for an int (FlagMetals)
    FlagMetals = binary_data.read(chunk_size)
    print(FlagAge)
    FlagMetals_convert = unpack('1i', FlagMetals)
    print('FlagMetals:', FlagMetals_convert, '\n')
    
    chunk_size = calcsize('6i')
    # 24 bytes for six ints (NallHW[6])
    NallHW = binary_data.read(chunk_size)
    print(NallHW)
    NallHW_convert = unpack('6i', NallHW)
    print('NallHW[6]:', NallHW_convert, '\n')
    
    chunk_size = calcsize('1i')
    # 4 bytes for an int (flag_entr_ics)
    flag_entr_ics = binary_data.read(chunk_size)
    print(flag_entr_ics)
    flag_entr_ics_convert = unpack('1i', flag_entr_ics)
    print('flag_entr_ics:', flag_entr_ics_convert, '\n')
    
    # UNUSED SPACE:
    
    chunk_size = calcsize('')
    # 4 bytes for an int (unused)
    unused = binary_data.read(chunk_size)
    print(unused)
    unused_convert = unpack('', unused)
    print('unused:', unused_convert, '\n')
    
    
    
```
