## Central black hole default mass value: 10000.0 (solar mass?)

```python
from bbh.sim import GadgetSnapshot, ParticleType
from physt import h1, h2, histogramdd

import numpy as np
import matplotlib.pyplot as plt
import physt

%matplotlib inline
```

```python
var = {}
for n in range(5):
    var[n] = "cbh_mass/s/snapshot_00%s" % n
```

```python
%%time
snap_list = {}
for snap in var:
    with open(var[snap], "rb") as binary_file:
        data = binary_file.read()
    snap_list[snap] = GadgetSnapshot()
    snap_list[snap].parse(data)
```

```python
print(var)
```

## For 1.0 CBH mass parameter:

```python
%%time
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
h2(snap_list[0].pos.T[0], snap_list[0].pos.T[1], 50*1, name="snap_000").plot("image", alpha=1, ax=axes[0], cmap="viridis", show_colorbar=True, cmap_normalize="log")
h2(snap_list[1].pos.T[0], snap_list[1].pos.T[1], 50*1, name="snap_001").plot("image", alpha=1, ax=axes[1], cmap="viridis", cmap_normalize="log")
h2(snap_list[2].pos.T[0], snap_list[2].pos.T[1], 50*10, name="snap_002").plot("image", alpha=1, ax=axes[2], cmap="viridis", cmap_normalize="log");

fig2, axes2 = plt.subplots(1, 2, figsize=(12, 4))
h2(snap_list[3].pos.T[0], snap_list[3].pos.T[1], 50*100, name="snap_003").plot("image", alpha=1, ax=axes2[0], cmap="viridis", cmap_normalize="log")
h2(snap_list[4].pos.T[0], snap_list[4].pos.T[1], 50*100, name="snap_004").plot("image", alpha=1, ax=axes2[1], cmap="viridis", cmap_normalize="log");
for ax in axes:
    ax.set_xlim(-50,50)
    ax.set_ylim(-50,50)
    ax.set_aspect(1)
for ax in axes2:
    ax.set_xlim(-50,50)
    ax.set_ylim(-50,50)
    ax.set_aspect(1)
```
